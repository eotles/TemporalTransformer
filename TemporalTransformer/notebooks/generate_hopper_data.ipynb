```python
import csv
import numpy as np
import math
```

```python
N = int(1E4)
id_form = '%05d'

n_jobs = int(N/10)
n_dxs = int(N)
```

```python
def log_decay_cat(N, n_cat_values=100, prefix='c'):
    n_cv = n_cat_values
    n_cvE = int(math.log10(n_cv))
    cv = ['%s%0{}d'.format(n_cvE) %(prefix, i) for i in range(n_cv)]
    p = np.diff(np.log(np.arange(n_cv+1)+1)/np.log(n_cv+1))
    cats = np.random.choice(cv, size=N, p=p)
    return(cats)
```

```python
ids = [id_form %(i) for i in range(N)]

ages = np.random.normal(45, 10, N)
ages = np.where(ages<18, 18, np.where(ages>70, 70, ages))
ages = np.round(ages)

sexs = np.random.choice(['F', 'M'], size=N)
eths = log_decay_cat(N, n_cat_values=6, prefix='e')
bnls = np.random.choice(['A', 'B', 'C'], size=N, p=[0.75, 0.249, 0.001])
jobs = log_decay_cat(N, n_cat_values=n_jobs, prefix='j')
```

```python
unique, counts = np.unique(jobs, return_counts=True)
for c,v in zip(unique, counts):
    print(c,v)
```

```python
def write_to_csv(fp, *cols):
    zipped_cols = zip(*cols)
    f = open(fp, "w")
    try:
        writer = csv.writer(f)   
        for row in zipped_cols:
            writer.writerow(row)
    finally:
        f.close()
```

```python
write_to_csv("data/characteristics_0.csv", 
             ids, ages, sexs, eths, bnls, jobs)
```
