```python
from Hopper import *
import tensorflow as tf
import numpy as np
import Prepper

import sys

%matplotlib inline
```

# Hopper

```python
verbose = True
h = dbms(verbose=verbose)
```

## Load Tables

```python
tc = table_config("characteristics_0", 
                  ["age", "sex", "eth", "bin_ldc", "job"], 
                  ["real", "bin", "ldc", "ldc", "hdc"],
                  has_times=False,
                  primary_key=True)
h.create_fvm_with_csv(tc, "data/characteristics_0.csv", delimiter=',')


_tc = table_config("samples_0", 
                  ["SBP", "DBP", "type"], 
                  ["real", "real", "ldc"], 
                  has_times=True,
                  primary_key=False)
h.create_fvm_with_csv(_tc, "data/samples_0.csv", delimiter=',')



_tc = table_config("samples_1", 
                  ["tmps", "hrs", "sbps", "rrs", "dbps", "sats", "wgts"],
                  ["real", "real", 'real', "real", "real", "real", "real"], 
                  has_times=True,
                  primary_key=False)
h.create_fvm_with_csv(_tc, "data/samples_1.csv", delimiter=',')


_tc = table_config("samples_2", 
                  ["dx"],
                  ["hdc"], 
                  has_times=True,
                  primary_key=False)
h.create_fvm_with_csv(_tc, "data/samples_2.csv", delimiter=',')



_tc = table_config("samples_3", 
                  ["outcome"],
                  ["bin"], 
                  has_times=True,
                  primary_key=False)
h.create_fvm_with_csv(_tc, "data/samples_3.csv", delimiter=',')


_tc = table_config("samples_4", 
                  ["q"],
                  ["bin"], 
                  has_times=True,
                  primary_key=False)
h.create_fvm_with_csv(_tc, "data/samples_4.csv", delimiter=',')
```

## Window, Filter, Transform, Aggregate, Normalize Data

```python
h.dew_it(after_first=35)
```

# Prepper 

```python
tfp = Prepper.tf_prepper(h)
tfp.features
```

## Fit Prepper to Desired Task

```python
tfp.fit(offsets=[0, 1, 2], label_fns=["samples_3/avg_outcome"], partition="train")
```

## Make Model

```python
r0 = tf.keras.layers.LSTM(units=32,return_sequences=True, name="RNN_0")
r1 = tf.keras.layers.LSTM(units=16,return_sequences=True, name="RNN_1")

final_model = tfp.build_model(middle_layer_list=[r0, r1])
final_model.compile(loss="binary_crossentropy")
final_model.summary()
```

## Make TensorFlow Dataset

```python
ds = tfp.transform_to_ds()
```

# Train Model

```python
#final_model.fit(ds["train"], epochs=1, validation_data=ds["dev"])
final_model.fit(ds["test"], epochs=2)
```

```python
def generate_comparison_sets():
    
    real = {lb: {o: [] for o in tfp.offsets} for lb in tfp.label_fns}
    pred = {lb: {o: [] for o in tfp.offsets} for lb in tfp.label_fns}
    
    for idx, prediction in predictions.items():
        label = tfp.labels[idx]
        
        for lb in tfp.label_fns:
            for o_i, o in enumerate(tfp.offsets):
                _y = [_[0] for _ in  label[lb][o_i]]
                real[lb][o] += _y
                
                _p = prediction[lb][:, o_i]
                _p = _p.tolist()
                pred[lb][o] += _p

    
    return(real, pred)
```

```python
from sklearn import metrics, calibration
import matplotlib.pyplot as plt
```

```python
def roc_curves(ys, ps, ns, coalate=True, lb=""): 
    plt.figure()
    plt.plot([0, 1], [0, 1], "k:")
    
    label_template = "{} (AUC: {:.2f})"
    
    all_y = []
    all_p = []
    for y, p, n in zip(ys, ps, ns):
        all_y += y
        all_p += p
        fpr, tpr, thresholds = metrics.roc_curve(y, p)
        roc_auc = metrics.auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=label_template.format(n, roc_auc))
    
    if coalate:
        fpr, tpr, thresholds = metrics.roc_curve(all_y, all_p)
        roc_auc = metrics.auc(fpr, tpr)
        plt.plot(fpr, tpr, 'k', lw=2, label=label_template.format("All", roc_auc))
    
        
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('FPR (1-Specificity)')
    plt.ylabel('TPR (Sensitivity)')
    plt.title('%s\nReceiver Operating Characteristics' %(lb))
    plt.legend(loc="lower right")
    plt.show()
```

```python
def cc_helper(y, p, n, ax1, ax2, lb="", color=None):
    label_template = "{} (AUC: {:.2f})"
    
    pr = [round(_) for _ in p]
    
    clf_score = metrics.brier_score_loss(y, p, pos_label=max(y))
    print("%s (%s)" %(lb, n))
    print("\tBrier: %1.3f" % (clf_score))
    print("\tPrecision: %1.3f" % metrics.precision_score(y, pr))
    print("\tRecall: %1.3f" % metrics.recall_score(y, pr))
    print("\tF1: %1.3f\n" % metrics.f1_score(y, pr))
    
    fraction_of_positives, mean_predicted_value = \
        calibration.calibration_curve(y, p, n_bins=10)

    if color:
        ax1.plot(mean_predicted_value, fraction_of_positives, "s-",
                 color=color,
                 label=label_template.format(n, clf_score))

        ax2.hist(p, range=(0, 1), bins=10, label=n,
                 color=color,
                 histtype="step", lw=2)
    else:
        ax1.plot(mean_predicted_value, fraction_of_positives, "s-",
                 label=label_template.format(n, clf_score))

        ax2.hist(p, range=(0, 1), bins=10, label=n,
                 histtype="step", lw=2)

#https://scikit-learn.org/stable/auto_examples/calibration/plot_calibration_curve.html#sphx-glr-auto-examples-calibration-plot-calibration-curve-py
def calibration_curves(ys, ps, ns, coalate=True, lb=""):
    
    fig = plt.figure(figsize=(10, 10))
    ax1 = plt.subplot2grid((3, 1), (0, 0), rowspan=2)
    ax1.plot([0, 1], [0, 1], "k:")
    ax2 = plt.subplot2grid((3, 1), (2, 0))
    
    all_y = []
    all_p = []
    
    for y, p, n in zip(ys, ps, ns):
        all_y += y
        all_p += p
        cc_helper(y, p, n, ax1, ax2, lb=lb)
        
    if coalate:
        cc_helper(all_y, all_p, "All", ax1, ax2, lb=lb, color="k")

    ax1.set_ylabel("Fraction of positives")
    ax1.set_ylim([-0.05, 1.05])
    ax1.legend(loc="lower right")
    ax1.set_title('%s\nCalibration' %(lb))

    ax2.set_xlabel("Mean predicted value")
    ax2.set_ylabel("Count")
    ax2.legend(loc="upper center", ncol=2)

    plt.tight_layout()
```

```python
def run_graph_functions(real, pred, graph_fx, coalate=True):
    
    for lb in tfp.label_fns:
        ys = []
        ps = []
        ns = []
        for o in tfp.offsets:            
            ys.append(real[lb][o])
            ps.append(pred[lb][o])
            ns.append(o)
        
        graph_fx(ys, ps, ns, lb=lb, coalate=coalate)

run_graph_functions(real, pred, roc_curves)
run_graph_functions(real, pred, calibration_curves)
```
